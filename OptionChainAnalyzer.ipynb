```python
import pandas as pd
import numpy as np

def load_nse_option_chain(csv_file):

    # Read CSV (skip the CALLS/PUTS row)
    df = pd.read_csv(
        csv_file,
        skiprows=1,
        engine="python"
    )

    # Rename columns
    df.columns = [
        "EMPTY",
        "CALL_OI",
        "CALL_CHNG_OI",
        "CALL_VOLUME",
        "CALL_IV",
        "CALL_LTP",
        "CALL_CHNG",
        "CALL_BID_QTY",
        "CALL_BID",
        "CALL_ASK",
        "CALL_ASK_QTY",
        "STRIKE",
        "PUT_BID_QTY",
        "PUT_BID",
        "PUT_ASK",
        "PUT_ASK_QTY",
        "PUT_CHNG",
        "PUT_LTP",
        "PUT_IV",
        "PUT_VOLUME",
        "PUT_CHNG_OI",
        "PUT_OI",
        "EMPTY2"
    ]

    # Remove unused columns
    df = df.drop(columns=["EMPTY", "EMPTY2"], errors="ignore")

    # Replace "-" with NaN
    df = df.replace("-", np.nan)

    # Remove commas from numbers
    df = df.replace(",", "", regex=True)

    # Convert everything to numeric
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    return df
```

```python
# Replace '--' with NaN
df.replace("--", np.nan, inplace=True)

# Convert every column except STRIKE to numeric
for col in df.columns:
    if col != "STRIKE":
        df[col] = pd.to_numeric(df[col], errors="coerce")

# STRIKE should also be numeric
df["STRIKE"] = pd.to_numeric(df["STRIKE"])

display(df.head())
df.info()
```

```python
def market_summary(df):
    summary = {}

    # Total Open Interest
    summary["Total Call OI"] = df["CALL_OI"].sum()
    summary["Total Put OI"] = df["PUT_OI"].sum()

    # Put Call Ratio
    if summary["Total Call OI"] > 0:
        summary["PCR"] = summary["Total Put OI"] / summary["Total Call OI"]
    else:
        summary["PCR"] = None

    # Highest Call OI
    call_row = df.loc[df["CALL_OI"].idxmax()]
    summary["Highest Call OI"] = call_row["CALL_OI"]
    summary["Call Resistance"] = call_row["STRIKE"]

    # Highest Put OI
    put_row = df.loc[df["PUT_OI"].idxmax()]
    summary["Highest Put OI"] = put_row["PUT_OI"]
    summary["Put Support"] = put_row["STRIKE"]

    return summary
```

```python
summary = market_summary(df)

for k, v in summary.items():
    print(f"{k:20} : {v}")
```

```python
def estimate_atm(df):
    temp = df.copy()

    temp["DIFF"] = (temp["CALL_LTP"] - temp["PUT_LTP"]).abs()

    atm = temp.loc[temp["DIFF"].idxmin()]

    return atm["STRIKE"]
```

```python
print("Estimated ATM Strike:", estimate_atm(df))
```
